```python
import pandas as pd
```

```python
df = pd.read_csv('final_csv.csv')
```

```python
df
```

```python
dummies = pd.get_dummies(df.location)
```

```python
dummies.head()
```

```python
df = pd.concat([df, dummies.drop('other',axis = 'columns')], axis = 'columns')
```

```python
df
```

```python
df = df.drop(['location'], axis = 'columns')
df
```

```python
X = df.drop(['price'], axis = 'columns')
```

```python
X
```

```python
y = df.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
print(f'Accuracy of the model is: {round(lr_clf.score(X_test, y_test)*100,2)} %')
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits= 5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X,y, cv= cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_the_best_model_using_grid_search_cv(X,y):
    algos = {
        'linear_regression':{
            'model':LinearRegression(),
            'params': {
                'positive': [True, False],
                 'fit_intercept': [True, False]
            }
        },
        'lasso':{
            'model':Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision-tree-regressor':{
            'model':DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        },
        
    }

    final_scores = []

    cv = ShuffleSplit(n_splits=5, test_size = 0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv= cv, return_train_score = False)
        gs.fit(X,y)
        final_scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_})
    return pd.DataFrame(final_scores, columns=['model','best_score', 'best_params'])
        
        
find_the_best_model_using_grid_search_cv(X, y)
```

```python
import numpy as np
def predict_price(location, sqft, bath, bhk):
    loc_ind = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_ind >= 0:
        x[loc_ind] = 1

    return lr_clf.predict([x])[0]


```
