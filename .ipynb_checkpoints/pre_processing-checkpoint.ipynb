```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_values(val):
    values = val.split('-')
    if len(values) == 2:
        return (float(values[0]) + float(values[1])) / 2 
    try:
        return float(val)
    except:
        return None
    
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_values)
```
